```python
import pandas as pd
df = pd.read_csv('C:/Project Entri/ipl_dataset.csv.csv')

print("First 5 rows of the dataset:")
display(df.head())

print("\nDataset Info:")
df.info()
print(f"\nShape of the dataset: {df.shape}")

print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nUnique values in each column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

print("\nStatistical summary:")
display(df.describe())
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Required colms
columns_to_keep = [
    'Innings','ID_x', 'Overs', 'BallNumber',
    'Batter', 'longName', 'Bowler', 'NonStriker',
    'BatsmanRun', 'ExtrasRun', 'TotalRun', 'IsWicketDelivery',
    'PlayerOut', 'Kind', 'FieldersInvolved',
    'BattingTeam', 'team1', 'team2', 'match_date',
    'toss_winner', 'toss_decision', 'winner',
    'player_of_match', 'venue', 'city'
]
```

```python
# Drop unwanted colms
columns_to_drop = [col for col in df.columns if col not in columns_to_keep]
df.drop(columns=columns_to_drop, inplace=True)
print("After dropping unwanted columns:", df.shape)
```

```python
# Standardizing text
text_columns = ['Batter', 'longName', 'Bowler', 'NonStriker', 'PlayerOut', 'Kind',
                'FieldersInvolved', 'BattingTeam', 'team1', 'team2', 'toss_winner',
                'toss_decision', 'winner', 'player_of_match', 'venue', 'city']

for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
    
# Standardize venue names
venue_map = {
    'Arun Jaitley Stadium, Delhi': 'Arun Jaitley Stadium',
    'Brabourne Stadium, Mumbai': 'Brabourne Stadium',
    'Dr Dy Patil Sports Academy, Mumbai': 'Dr Dy Patil Sports Academy',
    'Dr. Y.S. Rajasekhara Reddy Aca-Vdca Cricket Stadium, Visakhapatnam': 'Dr. Y.S. Rajasekhara Reddy Aca-Vdca Cricket Stadium',
    'Eden Gardens, Kolkata': 'Eden Gardens',
    'Feroz Shah Kotla': 'Arun Jaitley Stadium',
    'Himachal Pradesh Cricket Association Stadium, Dharamsala': 'Himachal Pradesh Cricket Association Stadium',
    'M Chinnaswamy Stadium, Bengaluru': 'M Chinnaswamy Stadium',
    'M.Chinnaswamy Stadium': 'M Chinnaswamy Stadium',
    'Ma Chidambaram Stadium, Chepauk': 'Ma Chidambaram Stadium',
    'Ma Chidambaram Stadium, Chepauk, Chennai': 'Ma Chidambaram Stadium',
    'Punjab Cricket Association Is Bindra Stadium': 'Punjab Cricket Association Stadium, Mohali',
    'Punjab Cricket Association Is Bindra Stadium, Mohali': 'Punjab Cricket Association Stadium, Mohali',
    'Punjab Cricket Association Is Bindra Stadium, Mohali, Chandigarh': 'Punjab Cricket Association Stadium, Mohali',
    'Rajiv Gandhi International Stadium, Uppal': 'Rajiv Gandhi International Stadium',
    'Rajiv Gandhi International Stadium, Uppal, Hyderabad': 'Rajiv Gandhi International Stadium',
    'Sawai Mansingh Stadium, Jaipur': 'Sawai Mansingh Stadium',
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium'
}

df['venue'] = df['venue'].replace(venue_map)
```

```python
# Normalizing team names
team_name_map = {
    'Rcb': 'Royal Challengers Bangalore',
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Mi': 'Mumbai Indians',
    'Dc': 'Delhi Capitals',
    'Dd': 'Delhi Daredevils',
    'Kxip': 'Punjab Kings',
    'Kings Xi Punjab': 'Punjab Kings',
    'Pbks': 'Punjab Kings',
    'Srh': 'Sunrisers Hyderabad',
    'Rr': 'Rajasthan Royals',
    'Kkr': 'Kolkata Knight Riders',
    'Csk': 'Chennai Super Kings',
    'Gt': 'Gujarat Titans',
    'Gl': 'Gujarat Lions',
    'Pune Warriors': 'Rising Pune Supergiant',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Rising Pune Supergiant': 'Rising Pune Supergiant',
    'Deccan Chargers': 'Deccan Chargers'
}

for col in ['BattingTeam', 'team1', 'team2', 'toss_winner', 'winner']:
    df[col] = df[col].replace(team_name_map)

# selecting only the active teams
active_teams = [
    'Chennai Super Kings', 'Delhi Capitals', 'Gujarat Titans',
    'Kolkata Knight Riders', 'Lucknow Super Giants', 'Mumbai Indians',
    'Punjab Kings', 'Rajasthan Royals', 'Royal Challengers Bangalore',
    'Sunrisers Hyderabad'
]

df = df[
    df['team1'].isin(active_teams) &
    df['team2'].isin(active_teams) &
    df['BattingTeam'].isin(active_teams) &
    df['toss_winner'].isin(active_teams) &
    df['winner'].isin(active_teams)
]
```

```python
# changing to date time
df['match_date'] = pd.to_datetime(df['match_date'], errors='coerce')
```

```python
# Dropping the missing rows
df.dropna(subset=['team1', 'team2', 'winner', 'player_of_match', 'BattingTeam'], inplace=True)
```

```python
# Replacing Nan
df.replace(['Nan', 'nan', 'NaN'], pd.NA, inplace=True)
df['PlayerOut'] = df['PlayerOut'].fillna('Not out')
df['Kind'] = df['Kind'].fillna('None')
df['FieldersInvolved'] = df['FieldersInvolved'].fillna('None')

# dropping rows of missinfg winner data
df.dropna(subset=['winner'], inplace=True)
df['longName'] = df['longName'].fillna('Unknown')
# dropping N values in the winner column
# Convert 'N' in 'winner' to NaN so it can be dropped
df['winner'].replace('N', pd.NA, inplace=True)
df.dropna(subset=['winner'], inplace=True)
#city missing values cahnging to unknown
df['city'] = df['city'].fillna('Unknown')

```

```python
# saving the data
df.to_csv("C:/Project Entri/Pre_processed_ipl_data.csv", index=False)
print(df.head())
print(df.isnull().sum())
print(df['team1'].unique())
```
